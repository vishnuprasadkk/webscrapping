### Importing relevant libraries

```python
from bs4 import BeautifulSoup
import requests
```

### Code

```python
rating=input('Enter a star rating "One", "Two", "Three", "Four", "Five" \n\n\n')
for i in range(1,51):
    html_script=requests.get(f'https://books.toscrape.com/catalogue/page-{i}.html').text
    soup=BeautifulSoup(html_script,'lxml')
    lst=soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
    print('********************************************************')
    print(f'The following books have {rating} star rating in page {i}\n\n\n')
    for item in lst:
        if item.p['class'][1]==rating:
            print(item.h3.a['title'])
            with open (f'Page{i}.txt', 'a', encoding="utf-8") as f:
                f.write(item.h3.a['title']+'\n')
    print('********************************************************')
    print('\n')
         
```
